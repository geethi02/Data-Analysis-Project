```python
import pandas as pd

# Load dataset
df = pd.read_excel('blinkit.xlsx')

# Clean column names
df.columns = (
    df.columns
    .str.strip()
    .str.replace(' ', '_')
)

# Calculations
total_items = df['Item_Type'].count()
avg_rating = df['Rating'].mean()

sales_by_outlet_size = df.groupby('Outlet_Size')['Sales'].sum()
sales_by_outlet_type = df.groupby('Outlet_Type')['Sales'].sum().sort_values(ascending=False)
top_5_items = df.groupby('Item_Type')['Sales'].sum().sort_values(ascending=False).head(5)
avg_rating_by_outlet = df.groupby('Outlet_Type')['Rating'].mean()

# Output
print("Total Items:", total_items)
print("\nAverage Rating:", avg_rating)

print("\nSales by Outlet Size:")
print(sales_by_outlet_size)

print("\nSales by Outlet Type:")
print(sales_by_outlet_type)

print("\nTop 5 Item Types by Sales:")
print(top_5_items)

print("\nAverage Rating by Outlet Type:")
print(avg_rating_by_outlet)
```

```python
import matplotlib.pyplot as plt

sales_by_outlet_type.plot(kind='bar', title='Sales by Outlet Type')
plt.ylabel('Sales')
plt.show()
```
